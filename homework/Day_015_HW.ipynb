# 1. 請用 numpy 建立一個 10 x 10, 數值分布自 -1.0 ~ 1.0 的矩陣並繪製 Heatmap

```python
"""
請用 numpy 建立一個 10 x 10, 數值分布自 -1.0 ~ 1.0 的矩陣並繪製 Heatmap
"""
matrix = x = np.random.uniform(-1,1,[10,10])
matrix
```

```python
plt.figure(figsize=(10,10))
"""
繪製 Heatmap
"""
cmap = sns.cubehelix_palette(start = 1, rot = 3, gamma=0.8, as_cmap = True)
heatmap = sns.heatmap(matrix, cmap = cmap, linewidths = 0.05)

plt.show()
```

# 2. 請用 numpy 建立一個 1000 x 3, 數值分布為 -1.0 ~ 1.0 的矩陣，並繪製 PairPlot (上半部為 scatter, 對角線為 hist, 下半部為 density)

```python
nrow = 1000
ncol = 3
"""
請用 numpy 建立一個 1000 x 3, 數值分布為 -1.0 ~ 1.0 的矩陣，
"""
matrix = x = np.random.uniform(-1,1,[1000,3])
matrix[:10]
```

```python
# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)
"""
並繪製 PairPlot (上半部為 scatter, 對角線為 hist, 下半部為 density)
Please replace "..." to correct plot function
"""
grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(sns.distplot)
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)

plt.show()
```


# 3. 請用 numpy 建立一個 1000 x 3, 數值分布為常態分佈的矩陣，並繪製 PairPlot (上半部為 scatter, 對角線為 hist, 下半部為 density)

```python
nrow = 1000
ncol = 3
"""
請用 numpy 建立一個 1000 x 3, 數值分布為常態分佈的矩陣，
"""
mu, sigma = 0, 0.1 # mean and standard deviation
matrix = np.random.normal(mu, sigma, [1000,3])
matrix[:10]
```

```python
# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)
"""
並繪製 PairPlot (上半部為 scatter, 對角線為 hist, 下半部為 density)
Please replace "..." to correct plot function
"""

grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(sns.distplot)
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)

plt.show()
```
